```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("uber.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df['pickup_longitude'].fillna(df["pickup_longitude"].mean(), inplace = True)
```

```python
df['pickup_latitude'].fillna(df["pickup_latitude"].mean(), inplace = True)
```

```python
df.head()
```

```python
df["pickup_datetime"] = pd.to_datetime(df["pickup_datetime"])
```

```python
df.info()
```

```python
df = df.assign(

    hour = df.pickup_datetime.dt.hour,
    day = df.pickup_datetime.dt.day,
    month = df.pickup_datetime.dt.month,
    year = df.pickup_datetime.dt.year,
    dayofweek = df.pickup_datetime.dt.dayofweek
)
```

```python
df.head()
```

```python
df.drop(["pickup_datetime"], axis =1, inplace = True)
df.head()
```

```python
df.drop(["Unnamed: 0", "key"], axis = 1, inplace = True)
```

```python
df.head()
```

```python
df = df[(df["pickup_longitude"] >= -90) & (df["pickup_longitude"] <= 90)]
df = df[(df["pickup_latitude"] >= -90) & (df["pickup_latitude"] <= 90)]
df = df[(df["dropoff_longitude"] >= -90) & (df["dropoff_longitude"] <= 90)]
df = df[(df["dropoff_latitude"] >= -90) & (df["dropoff_latitude"] <= 90)]
```

```python
df.info()
```

```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5*iqr
upper_bound = q3 + 1.5*iqr

outliers = ((df < lower_bound) | (df > upper_bound)).any(axis = 1)

df_no_outlier = df[~outliers]
df_no_outlier.head()
```

```python
df_corr = df_no_outlier.corr()
```

```python
sns.heatmap(data = df_corr)
```

```python
df_no_outlier.head()
```

```python
df_no_outlier.plot(kind = "box", subplots = True, layout = (7,2), figsize = (15, 20))
```

```python
df_no_outlier.head()
```

```python
import haversine as hs

travel_dist = []
for pos in range(len(df_no_outlier)):
    longi1, lati1, longi2, lati2 = df["pickup_longitude"].iloc[pos], df["pickup_latitude"].iloc[pos], df["dropoff_longitude"].iloc[pos], df["dropoff_latitude"].iloc[pos]
    loc1 = (longi1, lati1)
    loc2 = (longi2, lati2)
    c = hs.haversine(loc1, loc2)
    travel_dist.append(c)
    
df_no_outlier["dist_km"] = travel_dist
df_no_outlier.head()
```

```python
X = df_no_outlier.drop(["fare_amount"], axis = 1)
Y = df_no_outlier["fare_amount"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score, mean_squared_error

print("r2 score: ", r2_score(y_test, y_pred))
```

```python
print("MSE: ",  mean_squared_error(y_test, y_pred))
```

```python
print("MSE: ", np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

print("r2 score: ", r2_score(y_test, y_pred))
```
