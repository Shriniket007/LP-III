```python
current_x = 2
# delta_x is an initial step size for the gradient descent algorithm, set to 1.
delta_x = 1
maximum_iterations = 10000
iteration_counter = 1
# learning rate
rate = 0.01 
# precision sets the threshold for stopping the optimization 
# when the change in x becomes smaller than this value, set to 0.000001.
precision = 0.000001
```

```python
#  gradient (slope)
def slope(x):
    return 2*(x+3)
def value_y(x):
    return (x+3)**2
y=[]
x=[]
y.append(value_y(current_x))
x.append(current_x)
```

```python
# while(delta_x>precision and iteration_counter<maximum_iterations):
#     previous_x = current_x
#     current_x = previous_x - delta_x * slope(previous_x)
#     y.append(value_y(current_x))
#     x.append(current_x)
#     delta_x = abs(previous_x-current_x)
#     print(iteration_counter+1)
#     iteration_counter+=1
#     print(current_x)
while delta_x > precision and iteration_counter < maximum_iterations:
    previous_x = current_x
    current_x = previous_x - rate * slope(previous_x)
    y.append(value_y(current_x))
    x.append(current_x)
    delta_x = abs(previous_x - current_x)
    print(f"Iteration {iteration_counter+1}")
    iteration_counter += 1
    print(f"X = {current_x}")

print(f"Local Minima occurs at: {current_x}")
```
