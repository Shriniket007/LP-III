```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Churn_Modelling.csv')
df.head()
```

```python
df.drop(["RowNumber","CustomerId","Surname"],axis=1,inplace=True) #Dropping the unnecessary columns 
df.head()
```

```python
states = pd.get_dummies(df['Geography'])
states
```

```python
gender = pd.get_dummies(df['Gender'])
gender
```

```python
df = pd.concat([df,gender,states], axis = 1)
df.head()
```

```python
df.drop(['Geography','Gender'],axis=1,inplace=True)
```

```python
X = df[['CreditScore','Age','Tenure','Balance','NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary','Male','Germany','Spain']]
y = df['Exited']
```

```python
df.head()
```

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.20)
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
```

```python
X_train  = mms.fit_transform(X_train)
X_test = mms.transform(X_test)
```

```python
pip install tensorflow
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Initialize the model
model = Sequential()


# dense (fully connected) layer with 64 units and ReLU activation, input_dim specifies the input shape(no. of features)
# A layer with 64 units will have 64 individual neurons.
# Rectified Linear Unit, ReLU(x) = max(0, x)
# In other words, for any input value x,
# the output of the ReLU function is x if x is greater than or equal to 0, and it's 0 if x is less than 0
# logistic sigmoid, S-shaped curve that maps(squashes) input values to the range of (0, 1)

# Add layers to the model
model.add(Dense(units=64, activation='relu', input_dim=X_train.shape[1]))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))


# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# optimizer='adam': The Adam optimizer is a popular optimization algorithm.
# loss='binary_crossentropy': This is the loss function used for binary classification.
# metrics=['accuracy']: You want to track the accuracy of the model during training.

# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.1)

# epochs=50: The number of training epochs (iterations over the entire training dataset).
# batch_size=32: The number of samples used in each update of the model's weights.
# validation_split=0.1: A portion (10%) of the training data is used as a validation
# set to monitor model performance during training.
```

```python
pip install tensorflow
```

```python
model.summary()
```

```python
y_pred =model.predict(X_test)
y_pred = (y_pred > 0.5) #Predicting the result
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred),annot = True)
```

```python
print(classification_report(y_test,y_pred))
```
